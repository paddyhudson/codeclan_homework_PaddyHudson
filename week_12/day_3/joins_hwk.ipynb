```python
import pandas as pd
from janitor import clean_names
```

#1

```python
card = pd.read_csv('data/CardBase.csv').clean_names()
customer = pd.read_csv('data/CustomerBase.csv').clean_names()
fraud = pd.read_csv('data/FraudBase.csv').clean_names()
transaction = pd.read_csv('data/TransactionBase.csv').clean_names()
```

#2

```python
card_customer = card.merge(customer, how = 'left', on = 'cust_id')
card_customer
```

#3 

```python
fraud_transaction = fraud.merge(transaction, how = 'outer', on = 'transaction_id')
fraud_transaction
```

```python
hats = pd.read_csv('data/hat_observations.csv')
hats[['hat_colour', 'hat_type']] = hats.observation.str.split(pat = ",", expand = True)
hats
```
